```python
def citi_clean(label, month_number):
    '''This function is supposed to: import the monthly csv, turn each date 
    column into a date time object, extract the hour from the date time object,
    thn export the cleaned dataframe to a new csv in the /data folder'''
    #Import the csv
    getfile = "data/citibike_2016" + month_number + ".csv"
    output = "data/" + label + ".csv"
    label = pd.read_csv(getfile)
     #Rename the columns: 
    old_columns1 = [u'tripduration', u'starttime', u'stoptime', u'start station id',
       u'start station name', u'start station latitude',
       u'start station longitude', u'end station id', u'end station name',
       u'end station latitude', u'end station longitude', u'bikeid',
       u'usertype', u'birth year', u'gender']
    #New Coulumns
    new_columns = ['tripduration', 'starttime', \
        'stoptime', 'start_station id',
       'start_station_name', 'start_station_latitude',
       'start_station_longitude', 'end_station_id', 'end_station_name',
       'end_station_latitude', 'end_station_longitude', 'bikeid',
       'usertype', 'birth_year', 'gender']
    #Rename
    label.rename(columns=dict(zip(old_columns2, new_columns)), inplace=True)
    
    #change the datetime to datetime object
    label['starttime'] = pd.to_datetime(label['starttime'])
    label['stoptime'] = pd.to_datetime(label['stoptime'])
    label['start_hour'] = label.starttime.apply(lambda x: x.hour)
    label['end_hour'] = label.stoptime.apply(lambda x: x.hour)
    #export to csv
    label.to_csv(output)
    #label.to_file(output)
    return label
```

```python
def citi_cleannew(label, month_number):
    '''This function is supposed to: import the monthly csv, turn each date 
    column into a date time object, extract the hour from the date time object,
    thn export the cleaned dataframe to a new csv in the /data folder'''
    #Import the csv
    getfile = "data/citibike_2016" + month_number + ".csv"
    output = "data/" + label + ".csv"
    label = pd.read_csv(getfile)
    #Rename the columns: 
    old_columns2 = [u'Trip Duration', u'Start Time', u'Stop Time', u'Start Station ID',
       u'Start Station Name', u'Start Station Latitude',
       u'Start Station Longitude', u'End Station ID', u'End Station Name',
       u'End Station Latitude', u'End Station Longitude', u'Bike ID',
       u'User Type', u'Birth Year', u'Gender']
    #New Coulumns
    new_columns = ['tripduration', 'starttime', \
        'stoptime', 'start_station id',
       'start_station_name', 'start_station_latitude',
       'start_station_longitude', 'end_station_id', 'end_station_name',
       'end_station_latitude', 'end_station_longitude', 'bikeid',
       'usertype', 'birth_year', 'gender']
    #Rename
    label.rename(columns=dict(zip(old_columns2, new_columns)), inplace=True)
    #change the datetime to datetime object
    label['starttime'] = pd.to_datetime(label['starttime'])
    label['stoptime'] = pd.to_datetime(label['stoptime'])
    label['start_hour'] = label.starttime.apply(lambda x: x.hour)
    label['end_hour'] = label.stoptime.apply(lambda x: x.hour)
    #export to csv
    label.to_csv(output)
    return label
```
